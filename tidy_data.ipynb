## Cake Recipes



Read the cake_recipes table. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

```python
df_cake = pd.read_csv('untidy-data/cake_recipes.csv')
df_cake #so the numbers inside the chart are the ammount of taistyness
```

Tidy the data as necessary.

```python
cake_change = df_cake['recipe:position'].str.split(':',expand = True)
cake_change
```

```python
cake_change.columns = ['recipe', 'position']

cake_change
```

```python
df_cake = pd.concat([df_cake,cake_change], axis=1).drop(columns ='recipe:position')
df_cake
```

```python
df_cake = df_cake.melt(id_vars = ['recipe' , 'position'],var_name= 'temp', value_name= 'tastiness')
df_cake
```

Which recipe, on average, is the best? recipe b

```python
df_cake.groupby('recipe').tastiness.mean()
```

Which oven temperature, on average, produces the best results? 275

```python
df_cake.groupby('temp').tastiness.mean()
```

Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
df_cake.sort_values(by = 'tastiness').tail()
```

## Bonus: explore the other tables in the tidy_data database and reshape them as necessary so that they are in a tidy format.

```python
sales = pd.read_csv('untidy-data/sales.csv')
```

```python
sales_melt = sales.melt(id_vars='Product', var_name='year_and_measure')
sales_melt.head()
```

```python
year_and_measure_df = sales_melt.year_and_measure.str.split(' ',expand=True)
year_and_measure_df.columns= ['year','measure']
year_and_measure_df
```

```python
sales2 = pd.concat([sales_melt, year_and_measure_df], axis=1).drop(columns='year_and_measure')
sales2
```

```python
sales_tidy = sales2.pivot_table(index=['Product', 'year'], columns='measure', values='value')
sales_tidy
```

```python
sales_tidy.columns.name = ''
sales_tidy.reset_index(inplace=True)
sales_tidy
```
